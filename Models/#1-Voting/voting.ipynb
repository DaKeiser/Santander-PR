```python
res_2 = pd.read_csv("../input/righttovote/Voting/final_lag_2.csv")
res_3 = pd.read_csv("../input/righttovote/Voting/final_lag_3.csv")
res_4 = pd.read_csv("../input/righttovote/Voting/final_lag_4.csv")
res_5 = pd.read_csv("../input/sampled-according-to-time/sample_lag (1).csv")
```

```python
def split_row(row):
    return np.array(row['changed'].split(" "))
```

```python
products = ["product"+str(i) for i in range(5,0,-1)] #Naming products as per their importance, first product name product5 and so on..
products
```

```python
resultant_dfs = [res_2,res_3,res_4,res_5]
for r in resultant_dfs:
    split_rec = r.apply(split_row,axis=1)
    new_columns = []
    for i in range(5):
        new_columns.append([])
    for j in split_rec: #Number of best products predicted
        for i in range(5):
            new_columns[i].append(j[i])
    for i in range(5):
        r[products[i]] = new_columns[i] #adding the split 5 top products to same notebook
```

```python
#Populating all products into one df to use apply function later
big_df = pd.DataFrame()
name2 = [str(2)+p for p in products]
name3 = [str(3)+p for p in products]
name4 = [str(4)+p for p in products]
name5 = [str(5)+p for p in products]
big_df[name2] = res_2[products]
big_df[name3] = res_3[products]
big_df[name4] = res_4[products]
big_df[name5] = res_5[products]
big_df
```

```python
product_name = ['ind_ahor_fin_ult1','ind_aval_fin_ult1','ind_cco_fin_ult1','ind_cder_fin_ult1',\
               'ind_cno_fin_ult1','ind_ctju_fin_ult1','ind_ctma_fin_ult1','ind_ctop_fin_ult1',\
               'ind_ctpp_fin_ult1','ind_deco_fin_ult1','ind_deme_fin_ult1','ind_dela_fin_ult1',\
               'ind_ecue_fin_ult1','ind_fond_fin_ult1','ind_hip_fin_ult1','ind_plan_fin_ult1',\
               'ind_pres_fin_ult1','ind_reca_fin_ult1','ind_tjcr_fin_ult1','ind_valo_fin_ult1',\
               'ind_viv_fin_ult1','ind_nomina_ult1','ind_nom_pens_ult1','ind_recibo_ult1']
indices = dict(zip(product_name,[i for i in range(24)]))
weights_nb = {'2':0.4598,'3':0.4628,'4':0.4625,'5':0.4555,'6':0.4497} #Notebook original scores
rank_weights = {5:11, 4:7, 3:4, 2:2, 1:1} #Weighting the ordering the 5 products with most prominent product in each result getting more weight
column_names = list(big_df.columns)
def final_apply(row):
    weights = [0]*24
    for i in range(20):
        c = column_names[i]
        res_weight = weights_nb[str(c[0])]
        pos = int(c[-1])
        name = row[i]
        weights[indices[name]] += rank_weights[pos]*res_weight
    sort_index = np.argsort(weights)
    sort_index = sort_index[::-1]
    product_list = [product_name[k] for k in sort_index[:5]]
    recom_string = ' '.join(product_list)
    return recom_string
```

```python
final_apply(big_df.iloc[0]) #Testing the function with one row
```

```python
res = pd.DataFrame()
res['ncodpers'] = res_2['ncodpers']
res.head()
```

```python
res['changed'] = big_df.apply(final_apply,axis=1)
res.head()
```
